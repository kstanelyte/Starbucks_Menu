```python

full_menu.columns=full_menu.columns.str.lstrip()
full_menu.columns=full_menu.columns.str.rstrip()


drink_menu.columns=drink_menu.columns.str.lstrip()
drink_menu.columns=drink_menu.columns.str.rstrip()


food_menu.columns=food_menu.columns.str.lstrip()
food_menu.columns=food_menu.columns.str.rstrip()
food_menu.columns
```

#### Inspecting data in 'full_menu' columns:

```python
print(full_menu['Total Fat (g)'].unique()) 
print(full_menu['Vitamin C (% DV)'].unique()) 
print(full_menu['Vitamin A (% DV)'].unique())
print(full_menu['Calcium (% DV)'].unique()) 
print(full_menu['Iron (% DV)'].unique())
print(full_menu['Caffeine (mg)'].unique()) 
```

The following need to be dealt with:
* '3 2' in 'Total Fat (g)'
* '%' to be removed from 'Vitamin C (% DV)', 'Vitamin A (% DV)', 'Calcium (% DV)' and 'Iron (% DV)'
* '.00' to be removed from 'Iron (% DV)'
* Values of 'Varies', 'varies' and 'nan'

#### Finding which entry has 'Total Fat (g)' '3 2' value and change it to '3.2':

```python
full_menu.loc[full_menu['Total Fat (g)'] == '3 2']
```

```python
full_menu["Total Fat (g)"]=full_menu["Total Fat (g)"].str.replace('3 2', '3.2')
full_menu.loc[[237]]
```

#### Removing the '%' from % columns:

```python
full_menu = full_menu.replace({'Vitamin A (% DV)': {'%': ''}, 'Vitamin C (% DV)': {'%': ''},
                                     'Calcium (% DV)': {'%': ''}, 'Iron (% DV)': {'%': ''}}, regex=True)
```

#### Removing the '.00' from 'Iron (% DV)':

```python
full_menu = full_menu.replace({'Iron (% DV)': {'.00': ''}}, regex=True)
full_menu.head()
```

#### Checking for any n/a values in the data sets:

```python
print(full_menu.isna().any())

print(full_menu.isna().sum())
```

```python
drink_menu.replace('-', '', regex=True, inplace=True)
food_menu.replace('-', '', regex=True, inplace=True)
changed_columns = ['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein', 'Sodium']
drink_menu[changed_columns] = drink_menu[changed_columns].apply(pd.to_numeric)

print(drink_menu.dtypes)
print(food_menu.dtypes)
```

#### Correlation between different nutrients:

```python
sr1=full_menu.groupby('Beverage')['Total Fat (g)'].mean()
sr2=full_menu.groupby('Beverage')['Trans Fat (g)'].mean()
sr3=full_menu.groupby('Beverage')['Saturated Fat (g)'].mean()
sr4=full_menu.groupby('Beverage')['Total Carbohydrates (g)'].mean()
sr5=full_menu.groupby('Beverage')['Sugars (g)'].mean()
sr6=full_menu.groupby('Beverage')['Calories'].mean()
print('Total Fat vs. Calories =',(sr1.corr(sr6)))
print('Trans Fat vs. Calories =',(sr2.corr(sr6)))
print('Saturated Fat vs. Calories =',(sr3.corr(sr6)))
print('Total Carbohydrates vs. Calories =',(sr4.corr(sr6)))
print('Sugars vs. Calories =',(sr5.corr(sr6)))
```

* From a quick look at the relationship between the variables it can be seen that there is a positive correlation between all the nutrients that are usually associated with high calories.
* The highest correlation is observed between sugar content and calories as probably expected.
* Surprisingly, increase in saturated fat content does not result in increased calories for majority of drinks.

* Investigating why there's a weak correlation between calories and saturated fat content:

```python
full_menu.calories=full_menu[['Beverage', 'Beverage_prep','Calories', 'Saturated Fat (g)']]
full_menu.calories.sort_values(by=['Saturated Fat (g)'], ascending=False).head(10)
```

* Saturated fat value doesn't seem fluctuate much throughout the data set
* Highest saturated fat content doesn't come from the drinks highest in calories

```python
full_menu.calories=full_menu[['Beverage', 'Beverage_prep','Calories', 'Saturated Fat (g)']]
full_menu.calories.sort_values(by=['Saturated Fat (g)']).head(10)
```

* However, there's a clear decline in saturated fat content for beverages made with 'Soymilk', 'Nonfat Milk' or no milk at all when comparing to those prepared with '2% Milk'
